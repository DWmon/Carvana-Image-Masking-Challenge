```python
import os
import glob
from zipfile import ZipFile
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import keras
from keras.preprocessing import image
from sklearn.model_selection import train_test_split

from model.u_net import unet_128, unet_256, unet_512
from data_loader import data_generator
from evaluation import draw_plot
```

```python
from keras import backend as K
```

# Extract Data

```python
zip_path = 'E:\carvana_data\carvana-image-masking-challenge'
zip_list = glob.glob(zip_path+'\*.zip')
data_path = 'E:\carvana_data\data'
```

```python
%%time
for i in range(len(zip_list)):
    with ZipFile(zip_path + zip_list[i], 'r') as f: 
        f.extractall(data_path)
```

# Data load

```python
train_path = os.path.join(data_path, 'train')
train_list = os.listdir(train_path)
mask_path = os.path.join(data_path, 'train_masks')
mask_list = os.listdir(mask_path)
test_path = os.path.join(data_path, 'test')
test_list = os.listdir(test_path)

train_data = sorted(glob.glob(train_path + '\*.jpg'))
mask_data = sorted(glob.glob(mask_path + '\*.gif'))
test_data = sorted(glob.glob(test_path + '\*.jpg'))
```

```python
train_img, val_img, train_mask, val_mask = train_test_split(train_data, mask_data, test_size = 0.2, random_state = 1004)
print("Num of Train Data : " + str(len(train_img)))
print("Num of Validation Data : " + str(len(val_img)))
```

```python
fig_gen = data_generator(train_img, train_mask, 10, 128)
car_img, mask = next(fig_gen)

fig, ax = plt.subplots(10, 2, figsize=(10, 60))

for i in range(10):
    ax[i, 0].imshow(car_img[i].astype(np.float32))
    ax[i, 0].axis('off')
    ax[i, 0].title.set_text('Car')
    
    ax[i, 1].imshow(mask[i,:,:,0].astype(np.float32), cmap='gray')
    ax[i, 1].axis('off')
    ax[i, 1].title.set_text('Mask')

plt.show() 
```

# 1. Train & Test (U Net 128)

```python
img_size = 128
epochs = 100
batch_size = 16

train_gen = data_generator(train_img, train_mask, batch_size, img_size, train = True)
val_gen = data_generator(val_img, val_mask, batch_size, img_size)
model = unet_128()
model.summary()
```

```python
callbacks = [keras.callbacks.ReduceLROnPlateau(monitor = 'val_loss',
                                               factor = 0.8,
                                               patience = 2,
                                               verbose = 1, 
                                               min_lr = 0.00001),
            keras.callbacks.EarlyStopping(monitor = 'val_loss',
                                          patience = 4,
                                          verbose = 1,
                                          restore_best_weights = True),
            keras.callbacks.ModelCheckpoint(filepath = '.\weights\model%d\model%d_{epoch:03d}_{val_loss:.4f}.h5' %(img_size,img_size),
                                            monitor = 'val_loss',
                                            verbose = 1,
                                            save_best_only = True,
                                            save_weights_only = True),
            keras.callbacks.TensorBoard(log_dir='.\logs\model%d' %img_size,
                                        histogram_freq=0,
                                        write_graph=True,
                                        write_images=True)]
```

```python
hist = model.fit_generator(train_gen,
                          steps_per_epoch = len(train_img) / batch_size,
                          epochs = epochs,
                          validation_data = val_gen,
                          validation_steps = len(val_img) / batch_size,
                          callbacks = callbacks)
model.save_weights('./weights/model%d/model%d_best.h5' %(img_size, img_size))
```

```python
print("Train loss : ", hist.history['loss'][-1])
print("Dice coef : ", hist.history['dice_coef'][-1])
print("Validation loss : ", hist.history['val_loss'][-1])
print("Validation Dice coef : ", hist.history['val_dice_coef'][-1])
```

```python
draw_plot(hist)
```

```python
pred_gen = data_generator(train_img, train_mask, 10, img_size)
img, mask = next(pred_gen)
pred_mask = model.predict(img)
```

```python
fig, ax = plt.subplots(10, 3, figsize=(15,50))

for i in range(10):
    ax[i, 0].imshow(img[i].astype(np.float32))
    ax[i, 0].axis('off')
    ax[i, 0].title.set_text('Car')
    
    ax[i, 1].imshow(mask[i,:,:,0].astype(np.float32), cmap='gray')
    ax[i, 1].axis('off')
    ax[i, 1].title.set_text('Mask')
    
    ax[i, 2].imshow(pred_mask[i,:,:,0], cmap='gray')
    ax[i, 2].axis('off')
    ax[i, 2].title.set_text('Predicted Mask')
plt.show() 
```

```python
test_gen = data_generator(test_data,test_data, 10, img_size)
test_img, _ = next(test_gen)
test_mask = model.predict(test_img)
```

```python
fig, ax = plt.subplots(10, 2, figsize=(10,60))

for i in range(10):
    ax[i, 0].imshow(test_img[i].astype(np.float32))
    ax[i, 0].axis('off')
    ax[i, 0].title.set_text('Car')
 
    ax[i, 1].imshow(test_mask[i,:,:,0], cmap='gray')
    ax[i, 1].axis('off')
    ax[i, 1].title.set_text('Predicted Mask')
    
plt.show() 
```

```python
callbacks = [keras.callbacks.ReduceLROnPlateau(monitor = 'val_loss',
                                               factor = 0.8,
                                               patience = 2,
                                               verbose = 1, 
                                               min_lr = 0.00001),
            keras.callbacks.EarlyStopping(monitor = 'val_loss',
                                          patience = 3,
                                          verbose = 1,
                                          restore_best_weights = True),
            keras.callbacks.ModelCheckpoint(filepath = '.\weights\model%d\model%d_{epoch:03d}_{val_loss:.4f}.h5' %(img_size,img_size),
                                            monitor = 'val_loss',
                                            verbose = 1,
                                            save_best_only = True,
                                            save_weights_only = True),
            keras.callbacks.TensorBoard(log_dir='.\logs\model%d' %img_size,
                                        histogram_freq=0,
                                        write_graph=True,
                                        write_images=True)]
```

```python
hist = model.fit_generator(train_gen,
                          steps_per_epoch = len(train_img) / batch_size,
                          epochs = epochs,
                          validation_data = val_gen,
                          validation_steps = len(val_img) / batch_size,
                          callbacks = callbacks)
model.save_weights('./weights/model%d/model%d_best.h5' %(img_size, img_size))
```

```python
print("Train loss : ", hist.history['loss'][-1])
print("Dice coef : ", hist.history['dice_coef'][-1])
print("Validation loss : ", hist.history['val_loss'][-1])
print("Validation Dice coef : ", hist.history['val_dice_coef'][-1])
```
